# **MISSING DATA GUIDE**

```python
import DataLab as dl
import pandas as pd
import numpy as np
```

```python
df= dl.load_tabular('synth-data.csv')
df=dl.ColumnConverter(df, ['loan_amount']).to_numerical()
df.head()
```

```python
df.columns
```

```python
dl.MissingnessDiagnosis(df).show_missing_stats('percent')
```

```python
dl.Diagnosis(df).check_skewness()
```

```python
df[['credit_score', 'years_at_job', 'risk_score']] = dl.MissingHandler(df, ['credit_score', 'years_at_job', 'risk_score']).fill_with_mean()
```

```python
df[['age','income', 'expenses', 'savings', 'loan_amount','num_of_dependents']] = dl.MissingHandler(df, ['age','income', 'expenses', 'savings', 'loan_amount','num_of_dependents']).fill_with_median()
```

```python
df[['age','num_of_dependents', 'years_at_job']]=dl.DtypeConverter(df, ['age','num_of_dependents', 'years_at_job']).float64_to_int64()
```

```python
df=dl.NumericalCleaner(df).round_off(2)
```

```python
pd.options.display.float_format = '{:6f}'.format

data_max, data_min= df.max(), df.min()
n=len(df)

Range = data_max - data_min

bin_width = Range/n
bin_width
```

## **sturges**

```python
b=np.ceil(np.log2(n)+1)
Range/b
```

## **sqrt**

```python
Range/np.ceil(np.sqrt(n))
```

## **rice**

```python
Range/np.ceil(2 * n ** (1/3))
```

## **scott**


```python
3.5 * ((dl.Statistics(df).std_dev()) / (n ** (1/3)))
```

```python
(dl.Statistics(df).IQR()) / n ** (1/3)
```
